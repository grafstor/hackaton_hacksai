```python
def test_model(model_path, data_path):
    model = YOLO(model_path)
    results = model.val(data=data_path)
    return results

def print_results(results):
    print("Mean Average Precision (mAP) at IoU=0.50:", results.box.map50)         # mAP at IoU=0.50
    print("Mean Average Precision (mAP) at IoU=0.50:0.95:", results.box.map)      # mAP averaged across multiple IoU thresholds
    print("F1 Score:", results.box.f1)                                           # F1 score
    print("Precision:", results.box.p)                                           # Precision values
    print("Recall:", results.box.r)                                               # Recall values
```

#### From pretrained

```python
results = test_model("runs/detect/train3/weights/best.pt", "data.yaml")
print_results(results)
```

```python
results = test_model("runs/detect/train3/weights/best.pt", "benchmark.yaml")
print_results(results)                                            # Recall values
```
